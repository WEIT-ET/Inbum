1부터 10까지를 리스트에 삽입하는 코드를 작성해보자.



```python
arr=[]
for num in range(1,11):
  arr.append(num)

print(arr)
```

```python
arr2 = [x for x in range(1,11)]
print(arr2)
```

여기서 추가로 2의 배수만 넣고 싶어요!

```python
[x for x in range(1,11) if x%2 == 0]
```
