```python
import pandas as pd
```

**2. 데이터 프레임 만들기**





```python
df = pd.DataFrame({'name'    : ['김지훈', '이유진', '박동현', '김민지'],
                   'english' : [90, 80, 60, 70],
                   'math'    : [50, 60, 100, 20]})
df # 3개의 열 4개의 행
```

 열끼리 더 할 수 있다


```python
sum(df['english'])
```

```python
sum(df['math'])
```

```python
sum(df['math']) / 4     # 수학 점수 평균
```

```python
# 데이터 프레임의 행 개수 구하기
df = pd.DataFrame({'a' : [1, 2, 3],
                   'b' : [4, 5, 6]})
df
```
